```python
def load_jsonl_segment(file_path, start_line, num_lines=10):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for i, line in enumerate(file):
                if i >= start_line and i < start_line + num_lines:
                    print(line.strip())
                elif i >= start_line + num_lines:
                    break
    except Exception as e:
        print(f"Error loading file: {e}")

# 使用示例，查看第 20 行到第 30 行
load_jsonl_segment("datasets/ms_bench/ms_agent_bench_v1_sft_sample.jsonl", start_line=100)
```
